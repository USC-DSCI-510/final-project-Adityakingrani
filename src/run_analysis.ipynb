# Exploratory Data Analysis

```python
#Import required libraries
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
with open('../data/processed/team_mapping.pkl', 'rb') as file:
    team_map = pickle.load(file)
team_map
```

```python
with open('../data/processed/team_stats.pkl', 'rb') as file:
    team_stats = pickle.load(file)
team_stats
```

```python
#Reading our final dataframe using pandas 
df = pd.read_csv('../data/processed/data.csv').drop(columns=["Unnamed: 0"])
df
```

```python
# Analyzing First five rows of the df
df.head()
```

```python
# Basic statistics for t
df.describe()
```

```python
df.info()
```

```python
#Analyzing necessary columns
x = df.drop(columns=["winner"])
y = df[["winner"]]

display(x)
display(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')


# Train the model
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Test Accuracy: {accuracy:.2f}')

y_train_pred = model.predict(x_train)

# Calculate and print the training accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)

print(f'Train Accuracy:', train_accuracy)
```

```python
home_team = input("Enter home team: ")
away_team = input("Enter away team: ")

ip_df = {}
hid = team_map[home_team]
aid = team_map[away_team]


ip_df['home_agf'] = team_stats[hid]['agf']
ip_df['home_agc'] = team_stats[hid]['agc']
ip_df['home_aagf'] = team_stats[hid]['aagf']
ip_df['home_aagc'] = team_stats[hid]['aagc']
ip_df['h_pts'] = team_stats[hid]['pts']

ip_df['away_agf'] = team_stats[aid]['agf']
ip_df['away_agc'] = team_stats[aid]['agc']
ip_df['away_aagf'] = team_stats[aid]['aagf']
ip_df['away_aagc'] = team_stats[aid]['aagc']
ip_df['apts'] = team_stats[aid]['pts']
ip_df['home_team_id'] = hid
ip_df['away_team_id'] = aid

ip_df = pd.DataFrame(ip_df, index=[0])


res = model.predict(ip_df)
print(f'Winner: {home_team if res[0] == 1 else away_team}')
```
