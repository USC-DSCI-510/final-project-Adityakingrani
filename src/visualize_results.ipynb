# Data Vizualizatiion

```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/data.csv')
df
```

```python
df
```

```python
# Plotting a bar chart of the number of wins, draws, and losses
result_counts = df['winner'].value_counts()
result_counts.plot(kind='bar', color=['green', 'yellow', 'red'])
plt.title('Distribution of Match Results')
plt.xlabel('Results')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting a scatter plot for away goals for and against
plt.scatter(df['away_agf'], df['away_agc'], c=df['winner'], cmap='viridis')
plt.title('Away Goals For and Against')
plt.xlabel('Away Goals For')
plt.ylabel('Away Goals Conceded')
plt.colorbar(label='Result')
plt.show()
```

```python
# Plotting a scatter plot for home goals for and against
plt.scatter(df['home_agf'], df['home_agc'], c=df['winner'], cmap='coolwarm')
plt.title('Home Goals For and Against')
plt.xlabel('Home Goals For')
plt.ylabel('Home Goals Conceded')
plt.colorbar(label='Result')
plt.show()
```

```python
plt.hist(df['h_pts'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Home Points')
plt.xlabel('Home Points')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['apts'], bins=20, color='red', edgecolor='black')
plt.title('Distribution of Away Points')
plt.xlabel('Away Points')
plt.ylabel('Frequency')
plt.show()
```

```python
df.boxplot(column=['away_agf'], by='away_team_id', grid=False)
plt.title('Boxplot of Away Goals For by Team')
plt.suptitle('')  # Remove default title
plt.xlabel('Away Team ID')
plt.ylabel('Away Goals For')
plt.show()
```

```python
df.boxplot(column=['home_agf'], by='home_team_id', grid=False)
plt.title('Boxplot of Home Goals For by Team')
plt.suptitle('')  # Remove default title
plt.xlabel('Home Team ID')
plt.ylabel('Home Goals For')
plt.show()
```
