# DATA CLEANING AND PRE PROCESSING

```python
#Import required libraries
import pandas as pd
import pickle
```

```python
# Using .read_csv function by panadas to store all our data into different dataframes
all_matches_2021 = pd.read_csv('../data/raw/all_matches_2021.csv')
all_matches_2022 = pd.read_csv('../data/raw/all_matches_2022.csv')
all_matches_2023 = pd.read_csv('../data/raw/all_matches_2023.csv')

away_2021 = pd.read_csv('../data/raw/away_2021.csv')
away_2022 = pd.read_csv('../data/raw/away_2022.csv')
away_2023 = pd.read_csv('../data/raw/away_2023.csv')

home_2021 = pd.read_csv('../data/raw/home_2021.csv')
home_2022 = pd.read_csv('../data/raw/home_2022.csv')
home_2023 = pd.read_csv('../data/raw/home_2023.csv')
```

```python
# Removing null values and merging the dataframes
all_matches_2021 = all_matches_2021.dropna()
all_matches_2022 = all_matches_2022.dropna()
all_matches_2023 = all_matches_2023.dropna()
all_matches = pd.concat([all_matches_2021, all_matches_2022, all_matches_2023])
all_matches
```

```python
#Dropping unecessary columns
all_matches = all_matches[["home_team", "away_team", "score"]]
# WInner -> 1 if home 0 if away
def helper(x):
    if int(x.split(' - ')[0]) > int(x.split(' - ')[1]):
        return 1
    elif int(x.split(' - ')[0]) < int(x.split(' - ')[1]):
        return 0
    else:
        return 2
all_matches["winner"] = all_matches["score"].apply(lambda x: helper(x))
all_matches = all_matches.drop(columns=["score"])
all_matches
```

```python
home_df = pd.concat([home_2021, home_2022, home_2023])
away_df = pd.concat([away_2021, away_2022, away_2023])

home_df = home_df.drop(columns=["Unnamed: 0", "No"])
away_df = away_df.drop(columns=["Unnamed: 0", "No"])

display(home_df)
display(away_df)
```

```python
#Sorting df on the basis of points for each team
away_df = away_df.groupby("Name").sum().sort_values(by=["Pts"])[::-1]
home_df = home_df.groupby("Name").sum().sort_values(by=["Pts"])[::-1]
display(away_df)
display(home_df)
```

```python
#Merging all the dataframes using inner join
df = pd.merge(all_matches, home_df, left_on='home_team', right_on='Name', how='inner')
df
```

```python
#Calculating avaerage goals for and average goals conceded by the home team
df["home_agf"] = df["GF"] / df["GP"]
df["home_agc"] = df["GA"] / df["GP"]
df = df.rename(columns={"Pts": "home_hpts"}).drop(columns=["GP", "W", "D", "L", "GF", "GA", "GD"])
df
```

```python
#Renaming the required columns
df = pd.merge(df, away_df, left_on='home_team', right_on='Name', how='inner')
df["home_aagf"] = df["GF"] / df["GP"]
df["home_aagc"] = df["GA"] / df["GP"]
df = df.rename(columns={"Pts": "home_apts"}).drop(columns=["GP", "W", "D", "L", "GF", "GA", "GD"])
df
```

```python
df["h_pts"] = df["home_hpts"] + df["home_apts"]
df = df.drop(columns=["home_hpts", "home_apts"])
df
```

```python
#Merging all the dataframes to create our final df
df = pd.merge(df, home_df, left_on='away_team', right_on='Name', how='inner')
df["away_agf"] = df["GF"] / df["GP"]
df["away_agc"] = df["GA"] / df["GP"]
df = df.rename(columns={"Pts": "away_hpts"}).drop(columns=["GP", "W", "D", "L", "GF", "GA", "GD"])

df = pd.merge(df, away_df, left_on='away_team', right_on='Name', how='inner')
df["away_aagf"] = df["GF"] / df["GP"]
df["away_aagc"] = df["GA"] / df["GP"]
df = df.rename(columns={"Pts": "away_apts"}).drop(columns=["GP", "W", "D", "L", "GF", "GA", "GD"])
df
```

```python
df["apts"] = df["away_hpts"] + df["away_apts"]
df
```

```python
# Create a unique mapping for each team name
unique_teams = df['home_team'].unique()
team_mapping = {team: i+1 for i, team in enumerate(unique_teams)}

# Map the team names to unique numbers
df['home_team_id'] = df['home_team'].map(team_mapping)
df['away_team_id'] = df['away_team'].map(team_mapping)

df
```

```python
df = df.drop(columns=["home_team", "away_team", "away_hpts", "away_apts"])
df
```

```python
# Serialize the object to a binary string
team_map_obj = pickle.dumps(team_mapping)

# You can save the serialized data to a file if needed
with open('../data/processed/team_mapping.pkl', 'wb') as file:
    file.write(team_map_obj)
```

```python
#Saving our dataframe as a csv file
df.to_csv('../data/processed/data.csv')
```
